```python
import pandas as pd 
import streamlit as st 
import plotly.express as px 
import altair as alt
```

Column headers style is okay 

```python
df.isna().sum()
```

Here we see that we have a null objects in 5 columns:
* model_year 
* cylinders
* odometer
* paint_color 
* is_4wd  

So let's remove an empty values with "unknown" 

```python
columns_to_replace = ['model_year', 'cylinders', 'odometer', 'paint_color', 'is_4wd']

for column in columns_to_replace:
    df[column] = df[column].fillna('unknown')
```

```python
df.isna().sum() # looking for missing values one more time 
```

```python
df.duplicated().sum() # looking for a duplicates as well 
```

```python
df['model'].sort_values().unique()
```

Now we see that we got the clean data
